### Задание 1
В алгоритме $k-fold$ параметр $k$ влияет на качество выбора кипотез и на оценку true risk.
Увеличивая $k$, уменьшаем estimation error. Большой $k$ стоит выбирать в тех случаях, когда имеется достаточно априорных знаний о решаемой проблеме.
Уменьшая $k$, получим хуже обученную модель, но более точную оценку, т.е. лучше выбирать меньший $k$, когда важнее точность определения наилучших параметров алгоритма.

### Задание 2

Валидационная выборка обычно предназначена для подбора параметров модели, наиболее подходящих для решения задачи и минимизации $L_V$, тогда как на тестовой выборке оценивается $L_D$ и то, насколько выбранная модель с подобранными на валидационной выборке параметрами подходит для решения задачи.
При объединении тестовой и валидационной выборок можно получить неадекватную оценку пригодности модели для решения задачи.

### Задание 3

Недостатки алгоритма:

Leave-one-out практически пригоден только для маленьких выборок, т.к. количество разбиений выборки при возрастании размера выборки достигает неадекватно высоких значений.
Также недостатком алгоритма является тот факт, что в расчете оценки $L_D$ будут участвовать разбиения с очень малыми тренировочными или валидационными выборками, на которых  может быть получена совершенно неадекватная оценка.

Преимущества алгоритма:

Рассматривает не только подвыборки фиксированных размеров, но подвыборки всех возможных размеров, что позволяет заметить склонность алгоритма к переобучению (в этом случае на большом количестве разбиений будет получен плохой результат). Аналогично, хорошо обобщаемый алгоритм хорошо покажет себя на большем количестве  разбиений.
